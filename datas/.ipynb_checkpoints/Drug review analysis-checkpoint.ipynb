```python
# binary is obvious. let's separate the groups with another category 'satcat'
# Assumption: 2 and below is v.neg, 3 and up is ok to good
# these are two populations 
Drugs['Sat'] = np.nan
```

```python
Drugs.loc[Drugs.satisfaction <= 2, 'Sat'] = 'poor' 
Drugs.loc[Drugs.satisfaction == 3, 'Sat'] = 'ok'
Drugs.loc[Drugs.satisfaction > 3, 'Sat'] = 'good'
```

```python
Drugs.Sat.describe()
```

```python
DgroupB = Drugs.groupby(['Sat', 'drug'])['satisfaction'].agg(lambda x: len(x) / len(Drugs.loc['drug']))
DgroupB['count'].plot.bar()
```
